# Plot 2 signals in one figure

```python
def swap(x):
    list=[]
    first_element=x[:int(len(x)/2)]
    second_element=x[int(len(x)/2+1):]
    for element in second_element:
        list.append(element)
    for element in first_element:
        list.append(element)
    return np.array(list)
```

```python
plt.plot(x)
plt.plot(h)
plt.title("2 signals in one figure")
fig.set_size_inches(11,8)
plt.show()

```
